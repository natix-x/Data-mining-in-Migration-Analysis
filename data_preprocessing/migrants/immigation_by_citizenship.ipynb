```python
df = df.drop(columns=["STRUCTURE", "STRUCTURE_ID", "STRUCTURE_NAME", "Time", "Observation value", "OBS_FLAG", "Observation status (Flag) V2 structure", "CONF_STATUS", "Confidentiality status (flag)"])
```

```python
df = df.drop(columns=["sex", "Sex", "age", "Age definition", "Age class", "unit", "agedef", "freq", "Time frequency"])
```

```python
df = df.drop(columns=["geo", "citizen"])
```

```python
df.rename(columns={"Geopolitical entity (reporting)":"geo", "OBS_VALUE":"Migrants number", "TIME_PERIOD": "year"}, inplace=True)
```
