```python
import pandas as pd
```

**Population by educational attainment level, sex, age and citizenship (%)**
* Sex = Total
* Age = 15 - 64 
* Citizenship - foreigners
* Używana Międzynarodowa Standardowa Klasyfikacja Kształcenia (ISCED)
* Education level:
    * 'Less than primary, primary and lower secondary education (levels 0-2)' -> 'levels 0-2'
    * 'Upper secondary and post-secondary non-tertiary education (levels 3 and 4)' -> 'levels 3 and 4'
    * 'Tertiary education (levels 5-8)' -> 'levels 5-8'

```python
df = df = pd.read_csv('../../data/population_by_educational_attainment_level_and_citizenship.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[['International Standard Classification of Education (ISCED 2011)', 'Geopolitical entity (reporting)', 'TIME_PERIOD', 'OBS_VALUE']]
```

```python
df.head()
```

```python
df = df.rename(columns={
    'International Standard Classification of Education (ISCED 2011)': 'ISCED',
    'Geopolitical entity (reporting)': 'geo', 
    'TIME_PERIOD': 'year',
    'OBS_VALUE': 'Percentage'
})
```

```python
df['ISCED'].unique()
```

```python
def rename_education_levels(row):
    if row['ISCED'] == 'Less than primary, primary and lower secondary education (levels 0-2)':
        return 'levels 0-2'
    elif row['ISCED'] == 'Upper secondary and post-secondary non-tertiary education (levels 3 and 4)':
        return 'levels 3 and 4'
    elif row['ISCED'] == 'Tertiary education (levels 5-8)':
        return 'levels 5-8'

df['ISCED'] = df.apply(rename_education_levels, axis=1)
```

```python
df.info()
```

```python
for geo in df['geo'].unique():
    print(geo)
    df_geo = df[df['geo'] == geo]
    df_geo.info()
```
